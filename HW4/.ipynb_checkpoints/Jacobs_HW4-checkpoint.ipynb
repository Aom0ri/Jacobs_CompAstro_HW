```python
##Function to be integrated
def func(x):
    return 1/(x**(3/2))
##Rectangle rule
def MC_integrate(a,b,N):
    #Randomly interval of x's
    interval = np.random.uniform(low=a, high=b, size=N)
    val = 0
    for i in interval:
        val += func(i) #function value at current x
    I = ((b-a)/N) * val
    return I
```

```python
a = 1
b = 5
N = 10**4
MC_method = MC_integrate(a,b,N)
```

```python
print("The value of the integral with the this simple MC method with N = 10,000 is:", MC_method)
```
